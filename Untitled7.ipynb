```python
#spliting the dataset into independent and dependent variable
X = pd.DataFrame(np.c_[boston['LSTAT'], boston['RM']], columns = ['LSTAT','RM'])
Y = pd.DataFrame(np.c_[boston['MEDV']], columns = ['MEDV'])
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
#dividing the data into trainong set and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=5)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#linear regression is done for y=m1x1+m2x2+c
lin_model = LinearRegression()
#machine is trained on x_train and y_train data
lin_model.fit(X_train, Y_train)

#prediction is done on x_test data
x_train_predict = lin_model.predict(X_test)

print(x_train_predict)

```

```python
#comparison between the actual value of MEDV and the value prediced by machine
columns_tuple = tuple(Y_test.MEDV)
columns_tuple
df4 = pd.DataFrame(columns_tuple, columns =['Actual'])
df4['Predicted'] = x_train_predict
df4['difference']=df4['Actual']-df4['Predicted']
print(df4)
```

```python
#calculations of mean squared error and root mean squared error
mse= mean_squared_error(Y_test, x_train_predict)
rmse= np.sqrt(mean_squared_error(Y_test, x_train_predict))
print("The MSE value is:")
print(mse)
print("")
print("The RMSE value is:")
print(rmse)
```
