```python
import numpy as np
import nltk
from nltk.corpus import stopwords
```

```python
stringdata="I am Vinit Verulkar. I like to draw. Drawings are a very interesting tool to express your emotions. Using different colors in your drawing makes it more attractive and better. Drawings are very good in case of mental issues. "
stringdata2="I am Sonali Verulkar. I like to draw. Drawings are a very interesting tool to express your emotions. Using different colors in your drawing makes it more attractive and better. Drawings are very good in case of mental issues. "
stringdata3="Hello all good to see you all here . Hope you are doing well. I am also ok"
nltk_tokens = nltk.word_tokenize(stringdata)
nltk_tokens1 = nltk.word_tokenize(stringdata2)
nltk_tokens2 = nltk.word_tokenize(stringdata3)
print(nltk_tokens)
```

```python
stop_words = set(stopwords.words('english'))
nltk_tokens0 = [w for w in nltk_tokens if not w in stop_words]
nltk_tokens01 = [w for w in nltk_tokens1 if not w in stop_words]
nltk_tokens02 = [w for w in nltk_tokens2 if not w in stop_words]
print(nltk_tokens)
```

```python
tagged = nltk.pos_tag(nltk_tokens0)
#print(tagged)
nltk_tokens=[]
nltk_tokens1=[]
nltk_tokens2=[]
for word in tagged:
    if word[1]=='PRP':
        continue
    elif word[1]=='IN':
        continue
    elif word[1]=='RB':
        continue
    else:
        nltk_tokens.append(word[0])
    
tagged1 = nltk.pos_tag(nltk_tokens01)
#print(tagged)
for word in tagged1:
    #print(word[1])
    if word[1]=='PRP':
        continue
    elif word[1]=='IN':
        continue
    elif word[1]=='RB':
        continue
    else:
        nltk_tokens1.append(word[0])
    
tagged2 = nltk.pos_tag(nltk_tokens02)
#print(tagged)
for word in tagged2:
    if word[1]=='PRP':
        continue
    elif word[1]=='IN':
        continue
    elif word[1]=='RB':
        continue
    else:
        nltk_tokens2.append(word[0])
```

```python
nltk3=[]
for word in nltk_tokens:
    if(word=='.'):
        continue;
    else:
        nltk3.append(word)
nltk13=[]
for word in nltk_tokens1:
    if(word=='.'):
        continue;
    else:
        nltk13.append(word)
nltk23=[]
for word in nltk_tokens2:
    if(word=='.'):
        continue;
    else:
        nltk23.append(word)
print(tagged)
print(tagged1)
print(tagged2)
```

```python
from nltk.stem import WordNetLemmatizer
nltk2=[]
nltk12=[]
nltk22=[]
# Initialize wordnet lemmatizer
wnl = WordNetLemmatizer()
for word in nltk3:
    print ("{0:20}{1:20}".format(word, wnl.lemmatize(word, pos="v")))
    nltk2.append(wnl.lemmatize(word, pos="v"))
print(" ")
print(" ")
print(" ")
for word in nltk13:
    print ("{0:20}{1:20}".format(word, wnl.lemmatize(word, pos="v")))
    nltk12.append(wnl.lemmatize(word, pos="v"))
print(" ")
print(" ")
print(" ")
for word in nltk23:
    print ("{0:20}{1:20}".format(word, wnl.lemmatize(word, pos="v")))
    nltk22.append(wnl.lemmatize(word, pos="v"))
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer() #Initialize Python porter stemmer
nltk1=[]
nltk11=[]
nltk21=[]
for word in nltk2:
    print("{0:20}{1:20}".format(word, ps.stem(word)))
    word=ps.stem(word)
    nltk1.append(word)
print(" ")
print(" ")
print(" ")
for word in nltk12:
    print("{0:20}{1:20}".format(word, ps.stem(word)))
    word=ps.stem(word)
    nltk11.append(word)
print(" ")
print(" ")
print(" ")
for word in nltk22:
    print("{0:20}{1:20}".format(word, ps.stem(word)))
    word=ps.stem(word)
    nltk21.append(word)
```

```python
frequency = {}
frequency1 = {}
frequency2 = {}
listuniq=[]
listuniq1=[]
listuniq2=[]
count=0
count1=0
count2=0
for item in nltk1:
    count=count+1
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1
        listuniq.append(item)
for item in nltk11:
    count1=count1+1
    if item in frequency1:
        frequency1[item] += 1
    else:
        frequency1[item] = 1
        listuniq1.append(item)
for item in nltk21:
    count2=count2+1
    if item in frequency2:
        frequency2[item] += 1
    else:
        frequency2[item] = 1
        listuniq2.append(item)

print(frequency)
print(" ")
print(frequency1)
print(" ")
print(frequency2)
```

```python
print("TF")
print(" ")
for word in listuniq:
    print("{0:20}{1:20}".format(word, frequency[word]/ count))
print(" ")
print(" ")
print("TF")
print(" ")
for word in listuniq1:
    print("{0:20}{1:20}".format(word, frequency1[word]/ count1))
print(" ")
print(" ")
print("TF")
print(" ")
for word in listuniq2:
    print("{0:20}{1:20}".format(word, frequency2[word]/ count2))
```
