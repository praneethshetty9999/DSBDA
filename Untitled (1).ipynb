```python
import pandas as pd
import numpy as np
df= pd.read_csv("Iris.csv")
```

```python
df
```

```python
df.dropna()
```

```python
X = df.iloc[:, [1,2,3, 4]].values
Y = df.iloc[:, 5].values
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=5)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
# Fitting Naive Bayes to the Training set  
from sklearn.naive_bayes import GaussianNB  
classifier = GaussianNB()  
classifier.fit(X_train, Y_train)  
```

```python
Y_pred = classifier.predict(X_test)  
```

```python
print(Y_pred)
print(Y_test)
```

```python
columns_tuple = tuple(Y_test)
df4 = pd.DataFrame(columns_tuple, columns =['Actual'])
df4['Predicted'] = Y_pred
print(df4)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
TP1=cm[0][0]
FP1=cm[0][1]+cm[0][2]
FN1=cm[1][0]+cm[2][0]
TN1=cm[1][1]+cm[1][2]+cm[2][1]+cm[2][2]

TP2=cm[1][1]
FP2=cm[1][0]+cm[1][2]
FN2=cm[2][1]+cm[0][1]
TN2=cm[2][0]+cm[2][2]+cm[0][2]+cm[0][0]

TP3=cm[2][2]
FP3=cm[2][1]+cm[2][0]
FN3=cm[0][2]+cm[1][2]
TN3=cm[1][1]+cm[1][0]+cm[0][1]+cm[0][0]

print("The TP value for Iris-setosa is: ", TP1)
print("The FP value for Iris-setosa is: ",FP1)
print("The FN value for Iris-setosa is: ",FN1)
print("The TN value for Iris-setosa is: ",TN1)
print(" ")
print("The TP value for Iris-versicolor is: ", TP2)
print("The FP value for Iris-versicolor is: ",FP2)
print("The FN value for Iris-versicolor is: ",FN2)
print("The TN value for Iris-versicolor is: ",TN2)
print(" ")
print("The TP value for Iris-virginica is: ", TP3)
print("The FP value for Iris-virginica is: ",FP3)
print("The FN value for Iris-virginica is: ",FN3)
print("The TN value for Iris-virginica is: ",TN3)
recall1=TP1/(TP1+FN1)
#print("The Recall is: ",TP/(TP+FN))
precision1=TP1/(TP1+FP1)
#print("The Precision is: ",TP/(TP+FP))
recall2=TP2/(TP2+FN2)
#print("The Recall is: ",TP/(TP+FN))
precision2=TP2/(TP2+FP2)
recall3=TP3/(TP3+FN3)
#print("The Recall is: ",TP/(TP+FN))
precision3=TP3/(TP3+FP3)
print(" ")
print("species \t    ", "recall \t           ", "precission", "\t             F1 score")
print("Iris-setosa  \t    ", recall1, "\t \t \t", precision1, "\t \t \t",  (2*recall1*precision1)/(precision1+recall1))
print("Iris-versicolor  ", recall2, "\t", precision2, "\t",  (2*recall2*precision2)/(precision2+recall2))
print("Iris-virginica   ", recall3, "\t", precision3, "\t",  (2*recall3*precision3)/(precision3+recall3))
print(" ")
print("The macro average of f1-score is ",((2*recall1*precision1)/(precision1+recall1)+(2*recall2*precision2)/(precision2+recall2)+(2*recall3*precision3)/(precision3+recall3))/3)
print("The Accuracy of the model is: ",(TP1+TN1+TP2+TN2+TP3+TN3)/(TP1+TN1+FN1+FP1+TP2+TN2+FN2+FP2+TP3+TN3+FN3+FP3) )
print("The Error Rate is: ", 1-(TP1+TN1+TP2+TN2+TP3+TN3)/(TP1+TN1+FN1+FP1+TP2+TN2+FN2+FP2+TP3+TN3+FN3+FP3))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```
