```python
import pandas as pd
from sklearn import datasets
import seaborn as sns
df= pd.read_csv('train.csv')
df
```

```python
#The y axis will represent the sex of the passenger and the x axis will represent the age of the passenger
#the hue tells about the passenger survived or not
p= sns.boxplot(data=df, y="Sex", x="Age", hue="Survived") 
#styling of the representation is done specific name to the column, their font size, font weight is given 
p.set_xlabel('age', fontsize= 14, fontweight='bold')
p.set_ylabel('sex', fontsize= 14, fontweight='bold')
#title is give to the representation
p.set_title('Titanic data visualization', fontsize= 16,
fontweight='bold');
```
