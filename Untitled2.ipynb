```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("Personal.csv")
```

```python
df
```

```python
cate=['Schoolguy', 'Collegeguy', 'Jobguy']
df['Agecategory'] = pd.qcut(df['Age'], q=3, precision=0, labels=cate)
df
```

```python
mean= df.groupby(['Agecategory'])['Salary'].mean()
stddevi= df.groupby(['Agecategory'])['Salary'].std()
median= df.groupby(['Agecategory'])['Salary'].median()
mini= df.groupby(['Agecategory'])['Salary'].min()
maxi=df.groupby(['Agecategory'])['Salary'].max()
```

```python
print("Mean is: ")
print(mean)
print(" ")
print(" ")
print("Median is: ")
print(median)
print(" ")
print(" ")
print("Minimum is: ")
print(mini)
print(" ")
print(" ")
print("Maximum is: ")
print(maxi)
print(" ")
print(" ")
print("Standard Deviation is: ")
print(stddevi)
```

```python
out = df.groupby('Agecategory').agg({'Salary': [np.mean, np.median, np.min, np.max , np.std]})
out
```
